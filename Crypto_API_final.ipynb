```python
import requests
from datetime import datetime
import pandas as pd


def get_filename(from_symbol, to_symbol, datetime_interval, download_date):
    return '%s_%s_%s_%s_%s.csv' % (from_symbol, to_symbol, datetime_interval, download_date)


def download_data(from_symbol, to_symbol, datetime_interval):
    supported_intervals = {'minute', 'hour', 'day'}
    assert datetime_interval in supported_intervals,           'datetime_interval should be one of %s' % supported_intervals

    print('Downloading %s trading data for %s %s' %
          (datetime_interval, from_symbol, to_symbol))
    base_url = 'https://min-api.cryptocompare.com/data/histo'
    url = '%s%s' % (base_url, datetime_interval)
    print(url, "url ")
    params = {'fsym': from_symbol, 'tsym': to_symbol,
              'limit': 2000, 'aggregate': 1}
    request = requests.get(url, params=params)
    data = request.json()
    return data


def convert_to_dataframe(data):
    df = pd.io.json.json_normalize(data, ['Data'])
    df['datetime'] = pd.to_datetime(df.time, unit='s')
    df = df[['datetime', 'low', 'high', 'open',
             'close', 'volumefrom', 'volumeto']]
    return df


def filter_empty_datapoints(df):
    indices = df[df.sum(axis=1) == 0].index
    print('Filtering %d empty datapoints' % indices.shape[0])
    df = df.drop(indices)
    return df

def get_candlestick_width(datetime_interval):
      if datetime_interval == 'minute':
          return 30 * 60 * 1000  # half minute in ms
      elif datetime_interval == 'hour':
          return 0.5 * 60 * 60 * 1000  # half hour in ms
      elif datetime_interval == 'day':
          return 12 * 60 * 60 * 1000  # half day in ms
```

```python
## text processing from html
from textblob import TextBlob
import html2text
h = html2text.HTML2Text()
h.ignore_links = True
def process_html_body(df):
  if (len(df) == 0):
    return df
  for i in range(len(df["Body"])):
    df["Body"][i] = h.handle(df["Body"][i])
    df["Text"] = df["Headline"] + df["Body"]
  return df 

def sentiment_analysis(df):
  sentiment_score = []
  if (len(df) == 0):
    return [0,0]
  for desc in df["Text"]:
    analysis = TextBlob(desc)
    sentiment_score.append(analysis.sentiment)

  non_zero_sentiments = [ s for s in sentiment_score if s[0] != 0 and s[1]!=0] 
  if (len(non_zero_sentiments) == 0):
    return [0,0]

  avg_polarity = sum([ s[0] for s in non_zero_sentiments]) / len(non_zero_sentiments)
  avg_subjectivity  = sum([ s[1] for s in non_zero_sentiments]) / len(non_zero_sentiments)
  return [avg_polarity, avg_subjectivity]


def get_sentiment_scores(name, df):
  news_df = get_currency_news(name,df)
  news_df = process_html_body(news_df)
  return sentiment_analysis(news_df)


#get_sentiment_scores("Wrapped Bitcoin", merged_data_with_pairs)
```

```python
mar_cap_scoring = []
for mar_cap in merged_data_with_pairs['Market Capacity']:
  if mar_cap >= 500:
    mar_cap_scoring.append(90)
  elif mar_cap >= 250:
    mar_cap_scoring.append(85)
  elif mar_cap >= 75:
    mar_cap_scoring.append(75)
  elif mar_cap >= 50:
    mar_cap_scoring.append(65)
  elif mar_cap >= 30:
    mar_cap_scoring.append(55)
  elif mar_cap >= 15:
    mar_cap_scoring.append(45)
  else: 
    mar_cap_scoring.append(40)

sentiment_scoring = []
for sentiment in merged_data_with_pairs['Sentiment Avg']:
  if sentiment >= 0.5:
    sentiment_scoring.append(100)
  elif sentiment >= 0.4:
    sentiment_scoring.append(85)
  elif sentiment >= 0.3:
    sentiment_scoring.append(75)
  elif sentiment >= 0.2:
    sentiment_scoring.append(60) 
  elif sentiment >= 0.1:
    sentiment_scoring.append(45)
  elif sentiment >= 0:
    sentiment_scoring.append(30)
  else:
    sentiment_scoring.append(20)

change_scoring = []
for change in merged_data_with_pairs['Change %']:
  if change >= 1:
    change_scoring.append(100)
  elif change >= 0.5:
    change_scoring.append(85)
  elif change >= 0.2:
    change_scoring.append(75)
  elif change >= 0:
    change_scoring.append(60)
  elif change >= -0.5:
    change_scoring.append(45)
  elif change >= -1:
    change_scoring.append(30)
  else:
    change_scoring.append(20)
```
